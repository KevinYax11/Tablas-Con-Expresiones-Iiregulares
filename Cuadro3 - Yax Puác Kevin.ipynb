```python
from PyPDF2 import PdfReader
```

```python
rutaArch="/content/drive/MyDrive/Analisis Anual 2018 ETAS.pdf"
```

```python
documento = PdfReader (rutaArch)
```

```python
len(documento.pages)
```

```python
pagina = documento.pages[4]
texto = pagina.extract_text()
texto
```

```python
texto.split("\n")
```

```python
patron = r"\n(\d+)\s+(\d+)\s+([\w\s\/]+?)\s+([\w\s]+?)\s+([^0-9\n]+?)\s+(\d+)"
```

```python
datos= re.findall(patron,texto)
datos
```

```python
etas = pd.DataFrame(datos,columns=["No.","Se","Area de salud","ETA reportada", "Fuente de contagio","No. de casos"])
etas
```

```python
etas.dtypes
```

```python
etas["No."] = etas["No."].astype("int64")
etas["Se"] = etas["Se"].astype("int64")
etas["No. de casos"] = etas["No. de casos"].astype("int64")
etas.dtypes
```
